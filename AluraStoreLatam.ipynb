### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda_1 = tienda['Precio'].sum()
ingreso_total_tienda_2 = tienda2['Precio'].sum()
ingreso_total_tienda_3 = tienda3['Precio'].sum()
ingreso_total_tienda_4 = tienda4['Precio'].sum()

print(f"El ingreso total de la tienda 1 es: {ingreso_total_tienda_1}")
print(f"El ingreso total de la tienda 2 es: {ingreso_total_tienda_2}")
print(f"El ingreso total de la tienda 3 es: {ingreso_total_tienda_3}")
print(f"El ingreso total de la tienda 4 es: {ingreso_total_tienda_4}")
```

# 2. Ventas por categoría

```python
categoria1 = tienda['Categoría del Producto'].value_counts()
print(categoria1)
print(f"La categoría con más ventas de la tienda 1 es: {tienda['Categoría del Producto'].value_counts().idxmax()}")
```

```python
categoria2 = tienda2['Categoría del Producto'].value_counts()
print(categoria2)
print(f"La categoría con más ventas de la tienda 2 es: {tienda2['Categoría del Producto'].value_counts().idxmax()}")
```

```python
categoria3 = tienda3['Categoría del Producto'].value_counts()
print(categoria3)
print(f"La categoría con más ventas de la tienda 3 es: {tienda3['Categoría del Producto'].value_counts().idxmax()}")
```

```python
categoria4 = tienda4['Categoría del Producto'].value_counts()
print(categoria4)
print(f"La categoría con más ventas de la tienda 4 es: {tienda4['Categoría del Producto'].value_counts().idxmax()}")
```

# 3. Calificación promedio de la tienda


```python
calificacion = tienda['Calificación'].mean().round(2)
calificacion2 = tienda2['Calificación'].mean().round(2)
calificacion3 = tienda3['Calificación'].mean().round(2)
calificacion4 = tienda4['Calificación'].mean().round(2)

print(f"La calificación promedio de la tienda 1 es: {calificacion}")
print(f"La calificación promedio de la tienda 2 es: {calificacion2}")
print(f"La calificación promedio de la tienda 3 es: {calificacion3}")
print(f"La calificación promedio de la tienda 4 es: {calificacion4}")
```

# 4. Productos más y menos vendidos

```python
producto_mas1 = tienda['Producto'].value_counts().idxmax()
producto_menos1 = tienda['Producto'].value_counts().idxmin()

print(f"El producto más vendido de la tienda 1 es: {producto_mas1}")
print(f"El producto menos vendido de la tienda 1 es: {producto_menos1}")
```

```python
producto_mas2 = tienda2['Producto'].value_counts().idxmax()
producto_menos2 = tienda2['Producto'].value_counts().idxmin()

print(f"El producto más vendido de la tienda 2 es: {producto_mas2}")
print(f"El producto menos vendido de la tienda 2 es: {producto_menos2}")
```

```python
producto_mas3 = tienda3['Producto'].value_counts().idxmax()
producto_menos3 = tienda3['Producto'].value_counts().idxmin()

print(f"El producto más vendido de la tienda 3 es: {producto_mas3}")
print(f"El producto menos vendido de la tienda 3 es: {producto_menos3}")
```

```python
producto_mas4 = tienda4['Producto'].value_counts().idxmax()
producto_menos4 = tienda4['Producto'].value_counts().idxmin()

print(f"El producto más vendido de la tienda 4 es: {producto_mas4}")
print(f"El producto menos vendido de la tienda 4 es: {producto_menos4}")
```

# 5. Envío promedio por tienda

```python
envio = tienda['Costo de envío'].mean().round(2)
envio2 = tienda2['Costo de envío'].mean().round(2)
envio3 = tienda3['Costo de envío'].mean().round(2)
envio4 = tienda4['Costo de envío'].mean().round(2)

print(f'El gasto de envio promedio de la tienda 1 es {envio}')
print(f'El gasto de envio promedio de la tienda 2 es {envio2}')
print(f'El gasto de envio promedio de la tienda 3 es {envio3}')
print(f'El gasto de envio promedio de la tienda 4 es {envio4}')
```

### **Graficos**


```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_total_tienda_1, ingreso_total_tienda_2, ingreso_total_tienda_3, ingreso_total_tienda_4]

plt.bar(tiendas, ingresos)
plt.xlabel('Tiendas')
plt.ylabel('Ingresos')
plt.title('Ingresos por Tienda')
plt.show()
```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [calificacion, calificacion2, calificacion3, calificacion4]

plt.plot(tiendas, calificaciones)
plt.xlabel('Tiendas')
plt.ylabel('Calificaciones')
plt.title('Calificaciones por Tienda')
plt.show()
```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costo_de_envio = [envio, envio2, envio3, envio4]

plt.plot(tiendas, costo_de_envio)
plt.xlabel('Tiendas')
plt.ylabel('Costo de Envío')
plt.title('Costo de Envío por Tienda')
```
